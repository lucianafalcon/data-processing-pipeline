# An√°lisis Bivariado - Scatterplot (Dispersi√≥n)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x='PRECIO', y='PRECIO_TOTAL', data=df)
```

```python
sns.scatterplot(x='PRECIO', y='UNIDADES', data=df)
```
