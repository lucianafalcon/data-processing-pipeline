## Chequeo y limpieza de Nulos

```python
df.isnull()
```

```python
#Chequeo de Nulos
df.isnull().sum()
```

```python
#Otra Forma de Ver Nulos con una función
def nulos(df):
  cant = len(df)
  for col in df.columns:
    print(col,':',round(df[col].isnull().sum() * 100 / cant,2),'%')
nulos(df)
```
