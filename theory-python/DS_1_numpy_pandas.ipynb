Pandas - Dataframes

```python
# Pandas series es un array con indice
Numeros = range(50, 70, 2)
Numeros_serie = pd.Series(Numeros)
print(Numeros_serie)

print(Numeros_serie[2])
```

```python
Ajedrez_64 = np.arange(1,65).reshape(8,8)
Ajedrez_64,

Ajedrez_df = pd.DataFrame(
Ajedrez_64,
columns=range(1,9),
index=['A','B','C','D','E','F','G','H'])
Ajedrez_df
```

Dataframe - indexación

```python
#busca un unico dato desde el indice y la columna
Ajedrez_df.iloc[1,6]
```

```python
#busqueda de datos recorriendo indice por indice
Ajedrez_df.iloc[2:5]   #traeme desde el indice 2 al 5, las columnas
```

```python
#Busca por indice y diferencia por columnas
Ajedrez_df.loc['B':'F', [1,4,7]]  #traeme desde los indices B al F, las columnas 1,4 y 7
```

Armado de Datasets

```python
datos = {
    'Nombre': ['   Juan  ', 'María  ', '   Pedro', ' Ana  '],
    'Edad': [25, 30, 35, 40],
    'Correo': ['juan@example.com','Maria@sample.com', 'pedro@demo.com', 'ana@testing.com'],
    'Ciudad': ['Madrid', 'Barc.', 'Valencia', 'Sevilla']
}

dataframe = pd.DataFrame(datos)
print(dataframe)
```

Dataset - consulta de columna

```python
dataframe['Correo']
```

Dataset - crear una nueva columna

```python
dataframe['Pais'] = 'España'
dataframe
```

Dataset - manipulacion de datos (modificacion de los campos de la columna ciudad de minuscula a mayuscula)

```python
dataframe['Ciudad'] = dataframe['Ciudad'].str.upper()
dataframe
```

Dataset - agregado de una columna y rellenarlo con datos de otra columna existente

Ejemplo 1: se crea la columna Dominio con los datos de la columna correo correspondiente al dominio

```python
dataframe['Dominio'] = dataframe['Correo'].str.split('@').str[1]
dataframe
```

Ejemplo 2

```python
dataframe['Cod_ciudad'] = dataframe['Ciudad'].str[:3]
dataframe
```

Dataset - deteccion de patrones de Texto

```python
filtrado = dataframe[dataframe['Correo'].str.contains('example')]
#print(filtrado)
filtrado
```

Dataset - eliminación de espacios en blanco

```python
dataframe['Nombre'] = dataframe['Nombre'].str.strip()
print(dataframe)
```
